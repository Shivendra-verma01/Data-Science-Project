Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

Step 2 - Create the data frame

```python
dataframe=pd.read_csv("zomato_data (1).csv")
print(dataframe)
```

```python
dataframe
```

Convert the data type of column -rate

```python
def handleRate(value):
    value= str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu" ,fmt="d")
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()
```

Conclusion : Dinning resturants primarily accept offline orders ,whereas cafes primarily receive online orders. This suggests that clients
             Prefer orders in person at resturants, but prefer online ordering at cafes.
